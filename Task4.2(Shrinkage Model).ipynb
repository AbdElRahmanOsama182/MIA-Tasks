```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
data = pd.read_csv("shrink_ray_dataset.csv")
data.head()
```

```python
plt.scatter(data['Power'], data['Shrinkage'], color='red', label='Data Points',s=15)
plt.xlabel('Power (MW)')
plt.ylabel('Shrinkage Percentage (%)')
plt.title('Scatter Plot of the Dataset')
plt.legend()
plt.show()
```

```python
model = LinearRegression()
model.fit(data[['Power']], data['Shrinkage'])
```

```python
plt.scatter(data['Power'], data['Shrinkage'], color='blue', label='Data Points',s=15)
plt.plot(data['Power'], model.predict(data[['Power']]), color='red', label='Linear Regression')
plt.xlabel('Power (MW)')
plt.ylabel('Shrinkage Percentage (%)')
plt.title('Scatter Plot of the Dataset with Linear Regression')
plt.legend()
plt.show()
```

```python
knob = (85 - model.intercept_) / model.coef_
print(f"Knob setting for 85% shrinkage: {knob[0]} MW")
```
